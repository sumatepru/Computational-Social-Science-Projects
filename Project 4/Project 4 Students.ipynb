### Train PCA and Discuss Number of Components

```python
# fit PCA
# ----------
pca_all = PCA() # notice we do not specify the number of components if we want to keep them all
pca_all.fit(nhanes_scaled)
nhanes_scaled.shape
```

**I enter all 241 features into the PCA to leverage as much information as possible. PCA will organize that information into principal components for me, then I can choose how much information (the retained variance in PCAs) I want to use, given the trade-off is the model complexity.** 

### Barplot of Components

```python
# Create a Scree plot to evaluate the number of components to keep
# --------------------------------------------------


# create scree plot from PCA fit
# ----------
PC_values = np.arange(pca_all.n_components_) + 1
plt.plot(PC_values, pca_all.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

**From the above Scree Plot, the variance explained reduces so fast, just the third principal component variance is reduced to less-than-half of the first principal component, then the variance explained continues to decrease to 1% of the total variance at around the 20th principal component. After that, it asymptotically approaches zero.**

### How many components will you use to fit a supervised learning model?

```python
# assing the explained_variance_ratio from the PCA to a series to be plotted
explained_variance = pca_all.explained_variance_ratio_

# Plot the elbow curve
cumulative_variance = np.cumsum(explained_variance)
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Elbow Method for PCA')
plt.show()

# I choose 100 as the number of principle components
pca = PCA(n_components=100)
pca100 = pca.fit_transform(nhanes_scaled)
pca100.shape
```

**Based on the cumulative explained variance of PCA plot above, I choose 100 as the number of PCA because it can capture more than 80% of the total variance with 40% of the number of features. The decision is subjective - I see the marginal benefit of variance gained beyond this point is not worth the cost of model complexity - requiring more computing power.**

### 2D Scatterplot of the first two components

### Choose a Clustering Algorithm

**I choose the K-Means algorithm because it is simple and computationally efficient, so it works well with our dataset (large dataset) given the data is already standardized. Although the drawback is we must determine the number of clusters in the first place, the Elbow method can help us figure out the right number of clusters.**

### Cluster nhanes

```python
# Elbow method for K-Means
# Specify a range of k values
k_values = range(1, 20)

# Fit K-means models for each k and store the inertia in a list
inertia_values = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10, max_iter=300, random_state=42)
    kmeans.fit(nhanes_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow plot
plt.plot(k_values, inertia_values, marker='o')
plt.title('Elbow Plot for K-means Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

**Based on the Elbow plot, I choose K = 3. The inertia reduces significantly with K =2 and 3 - around 5% of the inertia each. After that when K = 4 the marginal reduction in inertia is around 2%, which is too small in my opinion. It is worth mentioning that the inertia remaining inertia is still high.**

```python
# specify kmeans algorithm
# ----------
kmeans = KMeans(n_clusters=3,
                n_init=10,
                max_iter=300 #default
                ).fit(nhanes_scaled)
```

### Plot your clusters on top of the BMI v. Income Poverty Ratio Plot

```python
# record the clusters
cluster_labels = kmeans.predict(nhanes_scaled)
nhanes_scaled_plot = nhanes_scaled
nhanes_scaled_plot['cluster_labels'] = cluster_labels

# plot in BMI and Income-Poverty Ration axises
ax = sns.scatterplot(x = "INDFMPIR", y = "BMXBMI", hue = "cluster_labels", palette = "tab10", data = nhanes_scaled_plot, s=10)
ax.set(xlabel = "Income to Poverty Line Ratio",
      ylabel = "Body Mass Index")
ax.set_title("BMI v. Income-Poverty Ratio")
plt.show()
```

## PCA and clustering address different aspects of understanding and analyzing datasets. PCA is primarily used for dimensionality reduction and feature extraction, aiming to represent the data in a more compact form. So it reduces the computational expense required to fit models and gives us a new perspective on how features link together through factor loadings in principle components. On the other hand, clustering is concerned with grouping similar data points together based on their inherent characteristics. ## Unsupervised learning techniques, such as clustering and PCA, can provide valuable insights and pre-processing advantages for downstream supervised learning tasks, especially in the context of big data, which is computationally expensive. For example, (1) big data often involves a large number of features, which can lead to the curse of dimensionality and increase computational complexity.
